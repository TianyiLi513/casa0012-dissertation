```python
pop = pd.read_excel('../data/sape23dt10amid2020coaunformattedsyoaestimateslondon.xlsx',
                        sheet_name='Mid-2020 Persons',header=4)
pop
# 人口结构参数计算，参考https://wenku.baidu.com/view/40d1e71dcaaedd3383c4d3d2.html?_wkts_=1675204057787&bdQuery=%E4%BA%BA%E5%8F%A3%E7%BB%93%E6%9E%84%E6%80%8E%E4%B9%88%E8%AE%A1%E7%AE%97
# 人口结构：性别比例，男性人数/女性人数*100
# 人口结构：年龄构成，按照参考的介绍，可分为少年、青年、老年三组的比例，这里为了简化，用年龄平均代替，其中90+的人员都设置为90岁

pop_male = pd.read_excel('../data/sape23dt10amid2020coaunformattedsyoaestimateslondon.xlsx',
                        sheet_name='Mid-2020 Males',header=4,usecols='A:C')
pop_female = pd.read_excel('../data/sape23dt10amid2020coaunformattedsyoaestimateslondon.xlsx',
                        sheet_name='Mid-2020 Females',header=4,usecols='A:C')

pop_male = pop_male.groupby('LSOA11CD')['All Ages'].sum()
pop_female = pop_female.groupby('LSOA11CD')['All Ages'].sum()

colls = pop.columns[2:]
pop2 = pd.DataFrame(pop.groupby('LSOA11CD')[colls].sum())

weight = np.array(range(0,91))
weight = np.expand_dims(weight,0).repeat(len(pop2),axis=0)


cc = pop2[colls[1:]]*weight
cc['sum'] = cc.apply(lambda x:sum(x),axis=1)
pop2['avg_age']=cc['sum']/pop2['All Ages']

pop2['gender_percent'] = pop_male/pop_female*100
pop = pop2
pop
```
