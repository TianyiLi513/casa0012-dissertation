```python
# -*- coding: utf-8 -*-
#import esda
import pandas as pd
import geopandas as gpd
from geopandas import GeoDataFrame
import libpysal as lps
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Point
#import contextily as ctx
from pylab import figure, scatter, show
import warnings
# warnings.filterwarnings('ignore')
# %matplotlib inline
from math import cos

#import geoplot as gplt
#import geoplot.crs as gcrs
from matplotlib.patches import Patch
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
from splot.esda import plot_moran,plot_moran_bv_simulation
import scipy.stats as ss   
from esda.moran import Moran,Moran_BV
```

```python
def find_code(point):
    for name, polygon in zip(df_map['GSS_CODE'], df_map['geometry']):
        if point.within(polygon):
            return name
    return ''
year = 2020 
```

```python
df_map = gpd.read_file('../data/statistical-gis-boundaries-london/ESRI', crs='ESRI:54004')

scattergdf = gpd.read_file(f'../result/data/osm_poi_data_selected/{year}/greater-london-{year}-free-selected.shp')
scattergdf = scattergdf.to_crs(crs='OSGB36 / British National Grid') #经纬度转笛卡尔坐标系
scattergdf['GSS_CODE'] = scattergdf['geometry'].apply(find_code)
scattergdf

temp = scattergdf['GSS_CODE'].value_counts()
temp = pd.DataFrame(temp)
temp.rename(columns={'GSS_CODE':'Retail_count'},inplace=True)
temp['GSS_CODE'] = temp.index
df_map2 = pd.merge(df_map,temp,left_on='GSS_CODE', right_on = 'GSS_CODE',
                   how='left')
df_map2 = df_map2.fillna(value={'Retail_count':0})
df_map2['Retail_Density']  = df_map2['Retail_count']/df_map2.area
df_map2
```

```python
#df_map_LSOA = gpd.read_file('../data/statistical-gis-boundaries-london/ESRI/LSOA_2011_London_gen_MHW.shp', crs='ESRI:54004')
df_map_LSOA = gpd.read_file(f'../result/data/{year}_LSOA_Retail_Density_map.shp')
df_map_LSOA
```

```python
pop = pd.read_excel('../data/sape23dt10amid2020coaunformattedsyoaestimateslondon.xlsx',
                        sheet_name='Mid-2020 Persons',header=4)
pop_male = pd.read_excel('../data/sape23dt10amid2020coaunformattedsyoaestimateslondon.xlsx',
                        sheet_name='Mid-2020 Males',header=4,usecols='A:C')
pop_female = pd.read_excel('../data/sape23dt10amid2020coaunformattedsyoaestimateslondon.xlsx',
                        sheet_name='Mid-2020 Females',header=4,usecols='A:C')
# 人口结构参数计算，参考https://wenku.baidu.com/view/40d1e71dcaaedd3383c4d3d2.html?_wkts_=1675204057787&bdQuery=%E4%BA%BA%E5%8F%A3%E7%BB%93%E6%9E%84%E6%80%8E%E4%B9%88%E8%AE%A1%E7%AE%97
# 人口结构：性别比例，男性人数/女性人数*100
# 人口结构：年龄构成，按照参考的介绍，可分为少年、青年、老年三组的比例，这里为了简化，用年龄平均代替，其中90+的人员都设置为90岁
```

```python
pop = pd.merge(pop,df_map_LSOA[['LAD11CD','LAD11NM','LSOA11CD']],left_on='LSOA11CD', right_on = 'LSOA11CD',
                   how='left')
pop
pop_male = pd.merge(pop_male,df_map_LSOA[['LAD11CD','LAD11NM','LSOA11CD']],left_on='LSOA11CD', right_on = 'LSOA11CD',
                   how='left')
pop_female = pd.merge(pop_female,df_map_LSOA[['LAD11CD','LAD11NM','LSOA11CD']],left_on='LSOA11CD', right_on = 'LSOA11CD',
                   how='left')
pop_female
```

```python
pop_male1 = pop_male.groupby('LAD11CD')['All Ages'].sum()
pop_female1 = pop_female.groupby('LAD11CD')['All Ages'].sum()
pop_female1
```

```python
colls = pop.columns[2:]
pop2 = pd.DataFrame(pop.groupby('LAD11CD')[colls].sum())
pop2
weight = np.array(range(0,91))
weight = np.expand_dims(weight,0).repeat(len(pop2),axis=0)

cc = pop2[colls[1:-2]]*weight
cc['sum'] = cc.apply(lambda x:sum(x),axis=1)
pop2['avg_age']=cc['sum']/pop2['All Ages']

pop2['gender_percent'] = pop_male1/pop_female1*100
pop3 = pop2[['All Ages','avg_age','gender_percent']]
pop3
```

```python
labels = ['all','1','2','3','4','5','6','7','8']

cols = ['Retail_cou', 'Retail_c_1', 'Retail_c_2', 'Retail_c_3',
       'Retail_c_4', 'Retail_c_5',  'Retail_c_6', 'Retail_c_7', 'Retail_c_8']

cols_d = ['Retail_den', 'Retail_d_1', 'Retail_d_2', 'Retail_d_3',
       'Retail_d_4', 'Retail_d_5',  'Retail_d_6', 'Retail_d_7', 'Retail_d_8']

Retail = df_map_LSOA.groupby('LAD11CD')[cols].sum()
Retail
```

```python
df_map3 = pd.merge(df_map2,pop3,left_on='GSS_CODE', right_on = 'LAD11CD',
                  how='left')
df_map3 = pd.merge(df_map3,Retail,left_on='GSS_CODE', right_on = 'LAD11CD',
                  how='left')
tips = ['Pop_Density','gender_percent','avg_age','GDHI']
df_map3[tips[0]] = df_map3['All Ages']/df_map3.area
df_map3 

gdhi = pd.read_excel('../data/regionalgrossdisposablehouseholdincomecityregions2020.xls',
                        sheet_name='Table 2',header=1)

for i,c in enumerate(cols):
    df_map3[cols_d[i]] = df_map3[cols[i]]/df_map3.area

df_map3 = pd.merge(df_map3,gdhi[['Geo code','2020']],left_on='GSS_CODE', right_on = 'Geo code',
                  how='left')
df_map3 = df_map3.rename(columns={'2020':'GDHI'})    
df_map3.to_file(f'../result/data/{year}_Borough_indexs.shp',
           driver='ESRI Shapefile',
           encoding='utf-8')
df_map3
```

```python

for t in tips:

    Moran_I_list = []
    random_Z_list = []
    random_Zp_list = []
    for i in range(0,len(cols)):
        x = df_map3[cols_d[i]]
        y = df_map3[t]
        wr = lps.weights.Rook.from_dataframe(df_map)
    
        moran = Moran_BV(x,y,wr)
        #moran.I
        
        # 输出Moran指数及其显著性检验结果
    
        Moran_I_list.append(moran.I)
        random_Z_list.append(moran.z_sim)
        random_Zp_list.append(moran.p_sim)
        
        # 双变量的图，可根据需要输出
        # fig, ax = plot_moran_bv_simulation(moran,  figsize=(10,10),aspect_equal=False)
        # plt.title(f'Moran plot for Retail Desity of LSOA for Retail Type {labels[i]} & Population')
        # path_out = '../result/plot/'
        # filename = f'Local_Moran_Retail_{labels[i]}_{labels[j]}'
        # plt.savefig(path_out+filename+'.png',
        #             dpi=500,bbox_inches = 'tight')
    
    Moran_I_df = pd.DataFrame()
    Moran_I_df['Retail'], Moran_I_df['Moran_I'], Moran_I_df['sim_Z'], Moran_I_df['sim_p'] = \
        (labels,Moran_I_list, random_Z_list, random_Zp_list)
    
    Moran_I_df.to_csv(f'../result/data/{year}_borough_MoranI_{t}.csv',index=False)
    
    # plot the scatter of population and Retail Density
    
    r_list = []
    p_list = []
    
    fig, axs = plt.subplots(3,3,figsize=(20,20))
    nT = 8
    for i in range(0,nT+1):
        ii = (i)//3
        jj = (i)%3
        
        x = df_map3[cols_d[i]]
        y = df_map3[t]
        r = ss.pearsonr(x,y)
        #print(f'correlation between Retail Density {labels[i]} & Population Density: {r[0]}')
        r_list.append(r[0])
        p_list.append(r[1])
        
        axs[ii,jj].scatter(x,y,marker='o',alpha=0.5)
        axs[ii,jj].set_title(f'Borough Retail Density {labels[i]}',fontsize=20)
        axs[ii,jj].set_xlabel(f'Density',fontsize=20)
        if(jj==0):
            axs[ii,jj].set_ylabel(t,fontsize=20)
        axs[ii,jj].xaxis.get_major_formatter().set_powerlimits((0,-5))
        
        pos_x = [0.25,0.52,0.8]
        pos_y = [0.83,0.57,0.3]
        plt.figtext(pos_x[jj], pos_y[ii],  f' r = {round(r[0],3)} \n p = {round(r[1],3)}', size=20)

    r_df = pd.DataFrame()
    r_df['Retail'], r_df['pearson correlation'], r_df['p']= \
        (labels,r_list, p_list)
    
    r_df.to_csv(f'../result/data/{year}_borough_correlation_{t}.csv',index=False)
    
    plt.suptitle(f'LSOA Retail Density & {t}',fontsize=40)
    path_out = '../result/plot/'
    filename = f'{year}_borough_Retail_Density_with_{t}'
    plt.savefig(path_out+filename+'.png',
                dpi=500,bbox_inches = 'tight')
```

```python
#多元线性回归
import statsmodels.api as sm

aaa = ['all','1','2','3','4','5','6','7','8']
fig, axs = plt.subplots(3,3,figsize=(20,20))

for iii,c in enumerate(cols_d):
    #print(c)
    x = sm.add_constant(data[tips]) #生成自变量
    y = data[c] #生成因变量
    model = sm.OLS(y, x) #生成模型
    result = model.fit() #模型拟合
    with open(f'../result/data/{year}_regression_summary_for_Retail_{aaa[iii]}.txt',
              'w') as f:
        print(result.summary(),file=f) #模型描述
    
    ii = (iii)//3
    jj = (iii)%3
    axs[ii,jj].axis('equal')
    axs[ii,jj].scatter(y,result.fittedvalues,marker='o',alpha=1)
    axs[ii,jj].set_title(f'Regression {aaa[iii]}',fontsize=20)
    axs[ii,jj].set_xlabel(f'Density Original',fontsize=20)
    if(jj==0):
        axs[ii,jj].set_ylabel(f'Density Regression',fontsize=20)
    axs[ii,jj].xaxis.get_major_formatter().set_powerlimits((0,-5))
    axs[ii,jj].yaxis.get_major_formatter().set_powerlimits((0,-5))
    
    axs[ii,jj].plot(y,y,'r-',linewidth=3,alpha=0.5)
    
    # pos_x = [0.25,0.52,0.8]
    # pos_y = [0.83,0.57,0.3]
    # plt.figtext(pos_x[jj], pos_y[ii],  f' r = {round(r[0],3)} \n p = {round(r[1],3)}', size=20)
        
plt.suptitle(f'Regression',fontsize=40)
path_out = '../result/plot/'
filename = f'{year}_regression'
plt.savefig(path_out+filename+'.png',
            dpi=500,bbox_inches = 'tight')
```
